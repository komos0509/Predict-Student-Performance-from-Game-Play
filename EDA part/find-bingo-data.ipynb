### 데이터 불러오기

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# Reference: https://www.kaggle.com/competitions/predict-student-performance-from-game-play/discussion/384359
dtypes={
    'elapsed_time':np.int32,
    'event_name':'category',
    'name':'category',
    'level':np.uint8,
    'room_coor_x':np.float32,
    'room_coor_y':np.float32,
    'screen_coor_x':np.float32,
    'screen_coor_y':np.float32,
    'hover_duration':np.float32,
    'text':'category',
    'fqid':'category',
    'room_fqid':'category',
    'text_fqid':'category',
    'fullscreen':'category',
    'hq':'category',
    'music':'category',
    'level_group':'category'}

dataset_df = pd.read_csv('/kaggle/input/predict-student-performance-from-game-play/train.csv', dtype=dtypes)
print("Full train dataset shape is {}".format(dataset_df.shape))
```

```python
dataset_1 = dataset_df[dataset_df['level_group'] == '0-4']
dataset_2 = dataset_df[dataset_df['level_group'] == '5-12']
dataset_3 = dataset_df[dataset_df['level_group'] == '13-22']
```

```python
dataset_1.reset_index()
dataset_2.reset_index()
dataset_3.reset_index()
```

```python
dataset_1_bingo = dataset_1[dataset_1['fqid'].str.contains('bingo', na=False)]
dataset_2_bingo = dataset_2[dataset_2['fqid'].str.contains('bingo', na=False)]
dataset_3_bingo = dataset_3[dataset_3['fqid'].str.contains('bingo', na=False)]
```
